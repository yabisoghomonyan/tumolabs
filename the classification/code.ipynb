```python
from scipy import stats
import plotly.express as px
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
df=pd.read_csv("/content/logatta.csv")
df.head()
```

```python
encoder=OrdinalEncoder()
for colums in df.columns:
  if df[colums].dtype=="object":
    df[colums]=encoder.fit_transform(df[[colums]])
df["accepted for the interview"]=encoder.fit_transform(df[["accepted for the interview"]])
df.head()
```

```python
stdsacle=StandardScaler()
for colums in df.columns:
  if df[colums].dtype=="int64":
    df[colums]=stdsacle.fit_transform(df[[colums]])
df.head()
```

```python
matrix = df.corr(numeric_only=True)
fig=px.imshow(matrix,x=matrix.columns,y=matrix.index,color_continuous_scale='RdBu_r',text_auto='.2f',aspect="auto")
fig.update_layout(title='Correlation Matrix',xaxis_showgrid=False,yaxis_showgrid=False,width=1000,height=1000,)
fig.show()
```

Compare to accepted for the interview , hightest corrolations have BiusinessTravell, Educatin and OverTime

```python
x_train,x_test,y_train,y_test=train_test_split(list(zip(df["BusinessTravel"],df["DailyRate"],df["Education"])),df["accepted for the interview"],test_size=0.2,random_state=42,shuffle=True)
dict_train={"BusinessTravel DailyRate Education":x_train,"accepted for the interview":y_train,"data":"Training"}
dict_test={"BusinessTravel DailyRate Education":x_test,"accepted for the interview":y_test,"data":"Test"}
df_all=pd.concat([pd.DataFrame(dict_train),pd.DataFrame(dict_test)],ignore_index=True)
df_all.head()
```

Logistic regression

```python
log_model=LogisticRegression()
log_model.fit(x_train,y_train)
accuracy = log_model.score(x_test, y_test)
print(f"Model Accuracy: {accuracy}")
predicted = log_model.predict([[33,1000,7]])
print(f"Prediction for input 40: {predicted}")
```

```python
y_pred=log_model.predict(x_test)
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

KNN

```python
knn_model=KNeighborsClassifier(n_neighbors=3)
knn_model.fit(x_train,y_train)
accuracy=knn_model.score(x_test,y_test)
print(f"Model Accuracy: {accuracy}")
predicted =knn_model.predict([[33,1000,7]])
print(f"Prediction for input 40: {predicted}")
```

```python
y_pred=knn_model.predict(x_test)
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

Gaussian Naive Bayes

```python
nb_model= GaussianNB()
nb_model.fit(x_train,y_train)
accuracy=nb_model.score(x_test,y_test)
print(f"Model Accuracy: {accuracy}")
predicted =knn_model.predict([[33,1000,77]])
print(f"Prediction for input 40: {predicted}")
```

```python
y_pred=nb_model.predict(x_test)
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```
